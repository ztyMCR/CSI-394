```python
import copy

n = 0
m = 0

i = 1
avg = 0
total = 0
zeroL = []
zeroOff = []
firL = []
secL = []
firOff = []
secOff = []
tempR = 0
tempG = 0
tempB = 0

with open('check.ppm', 'r') as fin:
  
  first_line = fin.readline()
  second_line = fin.readline()
  third_line = fin.readline()
  print ("The first line is "+ first_line)
  print ("The second line is " + second_line)
  print ("The third line is "+ third_line)
  width,height = second_line.split()
  print ("width is", width ," and height is", height)

  for line in fin:
    zeroL.append(int(line))
    
#done reading

rwvN = input("Please give the N for subtraction")
print("You entered: ", rwvN)


Ntimes  = int(rwvN)

while(Ntimes>0):
    
#average width and adding offset

    newWidth = int(int(width)/2)
    
    for num in zeroL:
       
        if i == 1:
            tempR = num
            i = i+1
            
        elif i == 2:
            tempG = num
            i = i+1
           
        elif i ==3:
            tempB = num
            i = i+1
        elif i == 4:
            avg = (tempR+num)/2
            firL.append(int(avg))
            firOff.append(int(tempR-int(avg)))
            i = i+1
        elif i == 5:
            avg = (tempG+num)/2
            firL.append(int(avg))
            firOff.append(int(tempG-int(avg)))
            i = i+1
        elif i == 6:
            avg = (tempB+num)/2
            firL.append(int(avg))
            firOff.append(int(tempB-int(avg)))
            i = 1
        n = n+1
        
        if n%int(768) == 0:
            print("the len of firL is ",len(firL))
            firL.extend(firOff)
            print("the len of firL is ",len(firL))
            print("the len of firOff is ",len(firOff))
            del firOff[:]
            print("merge!!! ",m," at n = ",n)
            
            m = m+1
    firL.extend(firOff)
        
    
    
     
    print("there are ",n," numbers")
    
    print("firstL has ",len(firL))
    print("width half picture done")

#average height and adding offsets
    
    newHeight = int(int(height)/2)
    newWidth = int(width)

    for i in range(newHeight):
        for j in range(newWidth):
            firstr2 = i*newWidth*3*2 +3*j
            secondr2 = i*newWidth*3*2 +3*j+3*newWidth
            firstg2 = firstr2+1
            secondg2 = secondr2+1
            firstb2 = firstg2+1
            secondb2 = secondg2+1
            avgr2 = int(firL[firstr2]+firL[secondr2])/2
            avgg2 = int(firL[firstg2]+firL[secondg2])/2
            avgb2 = int(firL[firstb2]+firL[secondb2])/2
            secL.append(avgr2)
            secL.append(avgg2)
            secL.append(avgb2)
#week3
            secOff.append(firL[firstr2]-avgr2)
            secOff.append(firL[firstg2]-avgg2)
            secOff.append(firL[firstb2]-avgb2)
            
        secL.extend(secOff)
        print("secOff has ", len(secOff))
        del secOff[:]
        
    print("secL has ",len(secL))
       
        
            
    
#reset vars
    del zeroL[:]
    zeroL = copy.deepcopy(secL)
    print("the new zeroL has ",len(zeroL)," elements")
    del firL[:]
    del secL[:]
    Ntimes = Ntimes -1
    
    width
    height 
    i = 1
    print("the new width ,height is",width," ",height)
    
    print("one loop done")
    


        
with open('checkRSPN.ppm', 'w') as fout:
    fout.write(first_line)
    fout.write(str(width))
    fout.write(" ")
    fout.write(str(height)+"\n")
    
    fout.write(third_line)
    for i in range(len(zeroL)):
        fout.write(str(int((zeroL[i])))+"\n")
        
print("done writing")
```
