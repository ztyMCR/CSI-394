```python
import copy

n = 0

i = 1
avg = 0
total = 0
firL = []
secL = []
tempR = 0
tempG = 0
tempB = 0
zeroL = []
with open('check1.ppm', 'r') as fin:
  
  first_line = fin.readline()
  second_line = fin.readline()
  third_line = fin.readline()
  print ("The first line is "+ first_line)
  print ("The second line is " + second_line)
  print ("The third line is "+ third_line)
  width,height = second_line.split()
  print ("width is", width ," and height is", height)

  for line in fin:
    zeroL.append(int(line))
    
#done reading

rwvN = input("Please give the N for subtraction")
print("You entered: ", rwvN)

Ntimes  = int(rwvN)
while(Ntimes>0):
    
    for num in zeroL:
      if i == 1:
        tempR = num
        i = i+1
      elif i == 2:
        tempG = num
        i = i+1
      elif i ==3:
        tempB = num
        i = i+1
      elif i == 4:
        avg = (tempR+num)/2
        firL.append(int(avg))
        i = i+1
      elif i == 5:
        avg = (tempG+num)/2
        firL.append(int(avg))
        i = i+1
      elif i == 6:
        avg = (tempB+num)/2
        firL.append(int(avg))
        i = 1
    
      #print (num)
      
      n = n+1  
    
    print("firstL has ",len(firL))
    print("width half picture done")

    newWidth = int(int(width)/2)
    newHeight = int(int(height)/2)

    for i in range(newHeight):
        for j in range(newWidth):
            firstr = i*newWidth*3*2 +3*j
            #print("firstr is ", firstr)
            secondr = i*newWidth*3*2 +3*j+3*newWidth
            #print("secondr is ", secondr)
            firstg = firstr+1
            secondg = secondr+1
            firstb = firstg+1
            secondb = secondg+1
            secL.append(int(firL[firstr]+firL[secondr])/2)
            secL.append(int(firL[firstg]+firL[secondg])/2)
            secL.append(int(firL[firstb]+firL[secondb])/2)
    
#reset vars
    del zeroL[:]
    zeroL = copy.deepcopy(secL)
    print("the new zeroL has ",len(zeroL)," elements")
    del firL[:]
    del secL[:]
    Ntimes = Ntimes -1
    width = newWidth
    height = newHeight
    i = 1
    print("the new width ,height is",width," ",height)
    
    print("one loop done")
    


        
with open('RSPN.ppm', 'w') as fout:
    fout.write(first_line)
    fout.write(str(newWidth))
    fout.write(" ")
    fout.write(str(newHeight)+"\n")
    
    fout.write(third_line)
    for i in range(len(zeroL)):
        fout.write(str(zeroL[i])+"\n")
        
print("done writing")
```
